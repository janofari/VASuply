```python
import pandas as pd
from sqlite_db_handler import insert_afectado

CSV_PATH = "/home/janofari/Documentos/proyectos/VASuply/afectados.csv"

# Mapeo de columnas del CSV a los campos esperados en la tabla 'afectados'
COLUMN_MAP = {
    "Día de alta": "dia_alta",
    "Afectado": "afectado",
    "Telefono": "tlf",
    "DNI": "dni",
    "Dirección afectada": "direccion_afectada",
    "Ubicacion alternativa": "ubi",
    "Población": "poblacion",
    "Situación personal": "situacion_personal",
    "Necesidad": "necesidad",
    "Día de visita": "dia_visita",
}

def load_csv():
    # Cargar CSV con pandas
    df = pd.read_csv(CSV_PATH, sep=",", encoding="utf-8")

    # Limpiar nombres de columnas por si hay espacios o caracteres raros
    df.columns = [col.strip() for col in df.columns]

    # Iterar por filas e insertar en SQLite
    for _, row in df.iterrows():
        data = {}
        for csv_col, db_col in COLUMN_MAP.items():
            data[db_col] = row.get(csv_col, None)

        # Validación simple por si hay NaN
        for key, value in data.items():
            if pd.isna(value):
                data[key] = None

        try:
            insert_afectado(data)
            print(f"✔ Insertado: {data['afectado']} ({data['dni']})")
        except Exception as e:
            print(f"❌ Error al insertar {data.get('afectado', 'desconocido')}: {e}")

load_csv()
```
